```python
from glob import glob

# !cd C:\proj\doctor\
img_list = glob('C:\\proj\\doctor\\data\\train\\images\\*')

print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/drive/MyDrive/doctor/data/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/drive/MyDrive/doctor/data/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('C:\\proj\\doctor\\data\\data.yaml', 'r') as f:
    data = yaml.safe_load(f)

print(data)

data['train'] = 'C:\\proj\\doctor\\data\\train.txt'
data['val'] = 'C:\\proj\\doctor\\data\\val.txt'

with open('C:\\proj\\doctor\\data\\data.yaml', 'w') as f:
    yaml.dump(data, f)

print(data)
```
